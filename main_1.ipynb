```python
import subprocess
def check_gcc_version(min_version=(4, 9)):
    try:
        # Run 'gcc --version' and get the output
        output = subprocess.run(["gcc", "--version"], capture_output=True, text=True, check=True)
        # Parse the first line of the output
        first_line = output.stdout.splitlines()[0]
        # Extract the version number (e.g., "4.9.3")
        version_str = first_line.split()[2]
        # Split version into major, minor, and patch
        version = tuple(map(int, version_str.split(".")[:2]))  # (major, minor)
        
        if version >= min_version:
            print(f"GCC version {version_str} is OK (>= {'.'.join(map(str, min_version))}).")
            return True
        else:
            print(f"GCC version {version_str} is too old (< {'.'.join(map(str, min_version))}).")
            return False

    except FileNotFoundError:
        print("GCC is not installed or not found in PATH.")
        return False
    except Exception as e:
        print(f"An error occurred while checking GCC version: {e}")
        return False
# Usage
if __name__ == "__main__":
    check_gcc_version()
```
